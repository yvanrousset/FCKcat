```python
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

with open(Path("..") / "data" / "results" / "all_results.json", 'r') as f:
    all_results = json.load(f)

models = [
    'ESM2v2_conc',
    'ESM2_max_poolingv2_conc',
    'ESM2_mixv2_conc',
    'ESM2_absmaxv2_conc',
    'ESM2_absmaxv2',
    'ESM2_absmaxv2_conc_reac',
    'rxnfp',
]
labels = [
    'mean pooling',
    'max pooling',
    'mix pooling',
    'absmax pooling\n(FCKcat)',
    r'$\delta$ absmax pooling',
    'absmax pooling\n+ rxn',
    'rxn',
]

metric = 'R2'

cv_data   = [all_results[m]['cv_results'][metric] for m in models]
test_data = [all_results[m]['test_results'][metric][0] for m in models]

fig, ax = plt.subplots(figsize=(14, 6))

bp = ax.boxplot(
    cv_data,
    positions=range(len(models)),
    widths=0.5,
    patch_artist=True,
    medianprops=dict(color='red', linewidth=2.5),
    boxprops=dict(facecolor='steelblue', alpha=0.5),
    whiskerprops=dict(linewidth=1.5),
    capprops=dict(linewidth=1.5),
    flierprops=dict(marker='o', markersize=4, linestyle='none'),
)

for i, vals in enumerate(cv_data):
    ax.scatter(i, np.mean(vals), color='black', zorder=5, s=60, marker='o')

for i, val in enumerate(test_data):
    ax.scatter(i, val, color='limegreen', zorder=6, s=80, marker='o')

ax.axvline(x=3.5, color='black', linewidth=1.2, linestyle='--')

ax.set_xticks(range(len(models)))
ax.set_xticklabels(labels, fontsize=12)
ax.set_ylabel(metric, fontsize=14 * 1.5)
ax.tick_params(axis='y', labelsize=12 * 1.5)
ax.set_xticklabels(labels, fontsize=12 * 1.2)

legend_handles = [
    mpatches.Patch(facecolor='steelblue', alpha=0.5, label='CV folds'),
    plt.Line2D([0], [0], color='red', linewidth=2.5, label='Median'),
    plt.Line2D([0], [0], marker='o', color='black', linestyle='None', markersize=8, label='Mean'),
    plt.Line2D([0], [0], marker='o', color='limegreen', linestyle='None', markersize=8, label='Test set'),
]
ax.legend(handles=legend_handles, fontsize=11 * 1.4, frameon=False, loc='upper right')

fig.tight_layout()
plt.show()
```
